### 요약 : 
##### 직원으로 뽑고 싶은 알바생 1위부터 7위와 
##### 인사담당자가 불쾌한 경험을 한 알바생 1위부터 6위
### 분석 :
#### 해당 기사에서 사용된 그래프는 hbar로 가장 뽑고싶은 알바생 1위가 한눈에 띄어 장점이있지만 전체적인 비율은 눈에 띄지 않는다. 그리고 인사팀이 알바생으로 인해 불쾌한 경험을 한 순위의 그래프는 없고 텍스트로만 설명이 되어있는 점이 좋지 않은것 같다.
### 결론 :
#### 그래서 파이차트로 해서 비율을 더 잘볼수 있게 만들어 가독성이 높아졌고, 뽑기 싫은 알바생 순위도 따로 파이차트로 만들어서 Best와 Worst를 구분하기 쉽게 만들었다. 하지만 단점은 1위가 눈에 띄지 않는다는 단점이 있다.


```python
import matplotlib.pyplot as plt

size = [39.0, 27.6, 11.5, 10.5, 5.6, 4.0, 1.5]
label = ["성실한 알바생","책임감이 투철한 알바생","오래 일할 수 있는 알바생", "밝고 명량한 알바생","열정이 넘치는 알바생","친절한 알바생","높은 성과를 내는 알바생"]        
color = ["steelblue","deepskyblue","darkturquoise","skyblue","lightskyblue","powderblue","aliceblue"]

size2 = [27.1,18.8,18.7,14.6,12.1,8.7]
label2 = ["말도 없이 안나오는 알바생","업무태도가 불량한 알바생","자주 지각하는 알바생","잦은 실수하는 알바생","일하다 도망가는 알바생","회사에 손해를 끼친 알바생"]
color2 = ["darkorange","goldenrod","tan","burlywood","bisque","oldlace"]

textprops1 = dict(weight=10,color="darkblue")
textprops2 = dict(weight=10,color="maroon")

plt.figure(figsize=[14,8])
plt.subplot(1, 2, 1)
plt.axis('equal')
plt.rc("font", family="Malgun Gothic")
plt.title("뽑고 싶은 알바생 TOP",fontsize=20, color="dodgerblue");
plt.pie(size, labels=label, autopct="%.1f%%", colors = color, explode=[0.05,0,0,0,0,0,0],labeldistance=0.9,textprops=textprops1)
plt.legend(loc=2,fontsize=9)

plt.subplot(1, 2, 2)
plt.axis('equal')
plt.rc("font", family="Malgun Gothic")
plt.title("뽑기 싫은 알바생 TOP",fontsize=20, color="orange");
plt.pie(size2, labels=label2, autopct="%.1f%%", colors = color2, explode=[0.05,0,0,0,0,0],labeldistance=0.9,textprops=textprops2)
plt.legend(loc=2,fontsize=9)

plt.show()
```
